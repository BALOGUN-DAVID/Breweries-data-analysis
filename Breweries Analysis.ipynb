```python
#Profit generated by each sales_rep
sns.countplot(y= "SALES_REP", data = breweries_data, hue_order = "PROFIT",orient = "p", saturation = 90)
```

```python
#profit across years
sns.boxenplot(x= "YEARS",y ="PROFIT", hue= "YEARS", data= breweries_data,outlier_prop = 0.02, orient = "V", dodge = 1)
```

```python
sns.violinplot(x= "REGION ", y = "PROFIT", data = breweries_data,palette = "tab10",gridsize =10, figsize = 20)
```

```python
#Region analysis

plt.figure(figsize=(15,10))
plt.subplot(2,3,1)
plt.xticks(rotation = 30)
sns.violinplot(x= "REGION ", y = "PROFIT", data = breweries_data,palette = "tab10",gridsize = 30,figsize = 30)


plt.subplot(2,3,2)
plt.xticks(rotation = 30)
sns.violinplot(x = "REGION ", y ="PLANT_COST", data = breweries_data, gridsize = 30, figsize = 30)


plt.subplot(2,3,3)
plt.xticks(rotation = 30)
sns.violinplot(x ="REGION ", y = "COSTS", data = breweries_data,palette = "PuBuGn", gridsize = 30, figsize = 30 )
plt.ylabel(' ')
plt.yticks([ ])

plt.subplot(2,3,4)
plt.xticks(rotation = 30)
sns.violinplot(x= "REGION ",y =  "QUANTITY", data = breweries_data, palette = "summer", gridsize = 30, figsize = 30)
plt.ylabel(' ')
plt.yticks([ ])

plt.subplot(2,3,5)
plt.xticks(rotation = 30)
sns.violinplot(x= "REGION ",y =  "UNIT_PRICE", data = breweries_data, palette = "summer", gridsize = 30, figsize = 30)
plt.ylabel(' ')
plt.yticks([ ])
```
